```python
df2=pd.read_csv("olist_order_items_dataset.csv")
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
df2['shipping_limit_date'] = pd.to_datetime(df2['shipping_limit_date'])
```

```python
#Create total value colum
df2['total_value'] = df2['price'] + df2['freight_value']
```

```python
#Create column for more detailed info
df2['order_month'] = df2['shipping_limit_date'].dt.month
df2['order_year'] = df2['shipping_limit_date'].dt.year
df2['order_day'] = df2['shipping_limit_date'].dt.day
df2['order_weekday'] = df2['shipping_limit_date'].dt.weekday  # 0=Mon, 6=Sun
```

```python
df2 = df2[df2['price'] > 0] 
```

```python
df2 = df2[df2['freight_value'] >= 0]
```

```python
df2.head()
```

```python
df2.to_csv("clean_olist_order_items_dataset.csv")
```
