```python
import pandas as pd
```

```python
# Orders
orders = pd.read_csv("Cleaned Dataset/clean_olist_orders_dataset.csv")

#Order Items
order_items = pd.read_csv("Cleaned Dataset/clean_olist_order_items_dataset.csv")

# Payments
payments = pd.read_csv("Cleaned Dataset/clean_olist_order_payments_dataset.csv")

# Reviews
reviews = pd.read_csv("Cleaned Dataset/clean_olist_order_reviews_dataset.csv")

# Products
products = pd.read_csv("Cleaned Dataset/clean_olist_products_dataset.csv")

# Sellers
sellers = pd.read_csv("Cleaned Dataset/cleaned_olist_sellers_dataset.csv")

# Geolocation
geolocation = pd.read_csv("Cleaned Dataset/cleaned_olist_geolocation_dataset.csv")

# Product Category Translation
product_category_translation = pd.read_csv("Cleaned Dataset/cleaned_product_category_name_translation.csv")

# Customers
customers = pd.read_csv("Cleaned Dataset/cleaned_olist_customers_dataset.csv")
```

```python
#Merge Product category
products = products.merge(
    product_category_translation,
    on="product_category_name",
    how="left"
)
```

```python
#Merge Order Items + Products
order_items_products = order_items.merge(
    products,
    on="product_id",
    how="left"
)
```

```python
#Merge Sellers
order_items_products = order_items_products.merge(
    sellers,
    on="seller_id",
    how="left"
)
```

```python
#Merge Orders (Customer + Delivery Info)
order_items_orders = order_items_products.merge(
    orders,
    on="order_id",
    how="left"
)
```

```python
#Merge Customers
order_items_orders = order_items_orders.merge(
    customers,
    left_on="customer_id",
    right_on="customer_id",
    how="left"
)
```

```python
#Merge ORDER ITEMS + PRODUCTS
order_items_products = order_items.merge(
    products,
    on="product_id",
    how="left"
)
```

```python
#Merge SELLER details
order_items_products = order_items_products.merge(
    sellers,
    on="seller_id",
    how="left"
)
```

```python
#Merge ORDERS (customer + delivery info
order_items_orders = order_items_products.merge(
    orders,
    on="order_id",
    how="left"
)
```

```python
#Merge CUSTOMERS
order_items_orders = order_items_orders.merge(
    customers,
    left_on="customer_id",
    right_on="customer_id",
    how="left"
)
```

```python
#Merge PAYMENTS
payment_agg = payments.groupby("order_id").agg(
    total_payment_value=("payment_value", "sum"),
    payment_installments=("payment_installments", "max"),
    zero_payment_flag=("is_zero_payment", "max")
).reset_index()

order_items_orders = order_items_orders.merge(
    payment_agg,
    on="order_id",
    how="left"
)
```

```python
#Merge REVIEWS
review_agg = reviews.groupby("order_id").agg(
    avg_review_score=("review_score", "mean"),
    has_review_text=("has_review_text", "max")
).reset_index()

final_df = order_items_orders.merge(
    review_agg,
    on="order_id",
    how="left"
)
```
